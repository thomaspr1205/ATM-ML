Image Preprocessing

```python
img_list = []

for dir in tqdm(os.listdir(os.getcwd())):
  for files in tqdm(os.listdir(os.getcwd() + f"/{dir}")):
    path = os.getcwd() + f"/{dir}/{files}/face_det_000000.bmp"
    if(os.path.isfile(path)):
      input_img = cv2.imread(path)
      input_img_resize=cv2.resize(input_img,(112,112))
      input_img_resize = input_img_resize / 225
      img_list.append(input_img_resize)
```

```python
img_data = np.array(img_list)
```

```python
img_data.shape
```

```python
num_classes = 2

num_of_samples = img_data.shape[0]
labels = np.ones((num_of_samples,),dtype='int64')
```

```python
import keras
from keras.utils import np_utils
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split

Y = np_utils.to_categorical(labels, num_classes)

#Shuffle the dataset
x,y = shuffle(img_data,Y, random_state=2)
# Split the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```
